```python
import pandas as pd
import numpy as np
```

### Karnataka


```python
acc_with_issues = pd.read_excel('ATC_MTD Dashboard_till Oct01-Oct29.xlsx',sheet_name=1)
acc_with_issues.drop(columns=['Normalized Account Number','Account Number','CNBD 2','Site Code'],inplace=True)
acc_with_issues
```

```python
acc_with_issue_final = acc_with_issues[(acc_with_issues.CircleName == "Karnataka")]
acc_with_issue_final 
```

```python
del_fail = pd.read_excel('ATC_MTD Dashboard_till Oct01-Oct29.xlsx',sheet_name=3)
del_fail.drop(columns=['StatementId','AccountNumber','SiteCode'],inplace=True)
del_fail
```

```python
del_fail_final = del_fail[(del_fail.Circle == "Karnataka")]
del_fail_final
```

```python
del_fail_final.shape
```

```python
def dfs_tabs(df_list, sheet_list, file_name):
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')   
    for dataframe, sheet in zip(df_list, sheet_list):
        dataframe.to_excel(writer, sheet_name=sheet, startrow=0 , startcol=0)   
    writer.save()


dfs = [acc_with_issue_final,del_fail_final]
sheets = ['Accounts with issues','Undelivered Accounts']    


dfs_tabs(dfs, sheets, 'Karnataka_Undelivered Accounts_Oct01-29.xlsx')
```
