### Delivery failures

```python
del_fail_sheet = pd.read_excel('OCT Json.xlsx',sheet_name=0)
del_fail_sheet.drop(columns=['StatementId','AccountNumber','SiteCode'],inplace=True)
del_fail_sheet
```

```python
del_fail_sheet["Statementdate"] = pd.to_datetime(del_fail_sheet["Statementdate"])  
date_range = (del_fail_sheet["Statementdate"] >= "2021-10-01") & (del_fail_sheet["Statementdate"] <= "2021-10-29") & (del_fail_sheet["Date"] >= "01.10.2021") & (del_fail_sheet["Date"] <= "29.10.2021")
temp_df1=del_fail_sheet.loc[date_range]
del_fail_final = temp_df1[['Date','ProviderAlias','Circle','Statementdate','DeliveryResult','ResponseCode','ResponseBody']]
del_fail_final
```

```python
del_fail_pivot = pd.pivot_table(del_fail_final, values='StatementId', index=['Circle','ProviderAlias'],
                    aggfunc={'StatementId': ['count']})
del_fail_pivot
```

```python
def dfs_tabs(df_list, sheet_list, file_name):
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')   
    for dataframe, sheet in zip(df_list, sheet_list):
        dataframe.to_excel(writer, sheet_name=sheet, startrow=0 , startcol=0)   
    writer.save()


dfs = [issue_acc_final,del_fail_final,del_fail_pivot,final_cust_df,cust_acc_table_final]
sheets = ['Accounts with issues','Delivery Failures',
          'delivery_failures_pivot','Customer portal accounts','cust_acc_portal_pivot']    


dfs_tabs(dfs, sheets, 'Final file using python.xlsx')
```
